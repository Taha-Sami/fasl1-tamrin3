```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x=np.array([[740,724,739],[724,739,727],[739,727,743],[727,743,736],[743,736,725],[736,725,731],[725,731,743],[731,743,740],[743,740,754],[740,754,775]])
```

```python
y=np.array([727,743,736,725,731,743,740,754,775,760]).reshape(-1,1)
```

```python
print(x.shape)
print(y.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
print(model.coef_,model.intercept_)
```

```python
y_predict=model.predict(x)
print(y)
print(y_predict)
```

```python
plt.plot(x[:,0],y,label='col1')
plt.plot(x[:,1],y,label='col2')
plt.plot(x[:,2],y,label='col3')
plt.legend()
```

```python
plt.plot(y,y_predict)
```
